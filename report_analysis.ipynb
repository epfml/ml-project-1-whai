```python
text=open("features_description.txt","r")
lines=text.readlines()
```

```python
lines=[line[:-1] for line in lines]
```

```python
by_cat=[]
for i in range(len(lines)):
    if "SAS Variable" in lines[i]:
        start=i
        break
    
for i in range(start+1,len(lines)):
    if "SAS Variable" in lines[i]:        
        by_cat.append(lines[start-2:i-2])
        start=i
```

```python
for i in range(len(by_cat)):
    if "of 137 A" in by_cat[i][-1]:
        by_cat[i]=by_cat[i][:-4]
```

```python
cat_infos={}
num_features=[]
for cat in by_cat:
    code=cat[2].split(" ")[-1]
    cat_infos[code]={"title":cat[0]}
    cat_infos[code]["description"]=cat[4]
    if "HIDDEN" in cat[8]:
            if cat[1].split(" ")[-1]!="Char":        
                cat_infos[code]["categorical"]=False
                num_features.append(code)
            else:
                cat_infos[code]["categorical"]=True
                cat_infos[code]["categories"]=None
            
    else:
        cat_infos[code]["categorical"]=True
        cat_infos[code]["categories"]={}
        # for a_value in cat[8:]:
        #     if ("")
        #     a_values=a_values.split(" ")
        #     cat_infos[code]["categories"]={a_values[0]:}
        
        
```
