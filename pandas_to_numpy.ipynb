```python
x_tr2=x_train.copy()
for col in intresting_features_indexes:
    # Create a mask for values to be replaced
    mask = np.isin(x_tr2[:, col], list(dico_transfos2[col].keys()))
    
    # Replace values in the column using a dictionary mapping
    x_tr2[:, col][mask] = [dico_transfos2[col].get(val, val) for val in x_tr2[:, col][mask]]
```

```python
x_tr_cleaned = x_tr2[:, intresting_features_indexes].copy()
fill_mean=False
if fill_mean:
    # Fill NaN values with the mean of each column
    for col in intresting_features_indexes:
        col_values = x_tr_cleaned[:, col]
        mean_value = np.nanmean(col_values)
        nan_mask = np.isnan(col_values)
        col_values[nan_mask] = mean_value
else:
    # Fill NaN values with 0
    nan_mask = np.isnan(x_tr_cleaned)
    x_tr_cleaned[nan_mask] = 0
```
